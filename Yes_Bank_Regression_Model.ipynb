### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

#### Hypothesis 1: Relationship Between Opening and Closing Prices

Statement:
There is a significant linear relationship between the opening price and the closing price of Yes Bank stock.

- Null Hypothesis (H₀): Open and Close prices are not linearly related.

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

#### Hypothesis 2: Monthly Seasonality in Closing Prices

Statement:
The average closing price of Yes Bank stock differs significantly across months.

- Null Hypothesis (H₀): Mean closing prices are equal across all months.

- Alternative Hypothesis (H₁): At least one month has a significantly different mean closing price.

Justification from Charts:

- Monthly boxplot shows variation in medians and spread

- Presence of seasonal volatility and outliers

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import f_oneway

# Ensure Date is datetime
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month

# Group closing prices by month
monthly_groups = [
    df[df['Month'] == m]['Close']
    for m in range(1, 13)
]

# Hypothesis Test 2
f_stat, p_value = f_oneway(*monthly_groups)

print("\nHypothesis Test 2: Monthly Closing Price Differences")
print(f"F-Statistic: {f_stat:.4f}")
print(f"P-Value: {p_value:.6f}")

if p_value < 0.05:
    print("Reject H0: Mean closing prices differ across months.")
else:
    print("Fail to Reject H0: No significant monthly difference.")
````

##### Which statistical test have you done to obtain P-Value?

- Statistical Test Used: One-Way ANOVA (Analysis of Variance)

- Why This Test:
It compares the mean closing prices across more than two independent groups (months).

- P-Value Interpretation:
Determines whether at least one month’s mean closing price differs significantly.

##### Why did you choose the specific statistical test?

Test Chosen: One-Way ANOVA

One-Way ANOVA was selected because the goal was to compare the mean closing prices across more than two independent groups (12 months). ANOVA efficiently determines whether observed differences in group means are statistically significant without inflating Type I error, which would occur if multiple pairwise tests were used.

##### Which text vectorization technique have you used and why?

The text vectorization technique primarily used in this code is **TF‑IDF Vectorization** implemented through **scikit‑learn**’s `TfidfVectorizer`. Although **CountVectorizer** is also demonstrated, TF-IDF is chosen as the recommended baseline for the final representation.

TF-IDF converts textual data into numerical feature vectors by calculating how important a word is in a document relative to the entire corpus. It multiplies **term frequency (TF)**, which measures how often a word appears in a document, with **inverse document frequency (IDF)**, which reduces the weight of common words that appear in many documents.

In this implementation, the vectorizer uses `max_features=1000` to limit the vocabulary size and `ngram_range=(1,2)` to capture both single words (unigrams) and two-word combinations (bigrams). Additionally, `stop_words='english'` removes common words that carry little meaning.

TF-IDF is preferred because it highlights informative terms, reduces noise from frequent words, and produces sparse matrices suitable for machine learning models and efficient storage.

````python
import pandas as pd
import numpy as np
from sklearn.linear_model import LassoCV
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
import logging

# ==============================
# STEP 1: Select Features
# ==============================

feature_columns = ['Open','High','Low']
target_column = 'Close'

X = df[feature_columns]
y = df[target_column]

# ==============================
# STEP 2: Handle Missing Values
# ==============================

X = X.fillna(X.mean())

# ==============================
# STEP 3: Scaling
# ==============================

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ==============================
# STEP 4: Lasso Feature Selection
# ==============================

lasso = LassoCV(cv=5, random_state=42)
lasso.fit(X_scaled, y)

lasso_selected = X.columns[lasso.coef_ != 0]

print("Features selected by Lasso:")
print(list(lasso_selected))

# ==============================
# STEP 5: Random Forest Importance
# ==============================

rf = RandomForestRegressor(random_state=42)
rf.fit(X, y)

importance = pd.Series(rf.feature_importances_, index=X.columns)
importance = importance.sort_values(ascending=False)

print("\nFeature Importance:")
print(importance)

# ==============================
# Final Features
# ==============================

final_features = list(X.columns)

print("\nFinal Selected Features:")
print(final_features)

X_final = X[final_features]

print("\nFeature selection completed successfully.")
````

##### What all feature selection methods have you used  and why?

Feature Selection Methods Used and Rationale
Removing Highly Correlated Features

Method: Calculated the absolute correlation matrix (X.corr().abs()) and removed features where the correlation coefficient with another feature was above a certain threshold (0.9 in this case).
Rationale: High correlation between independent variables (multicollinearity) can lead to unstable model coefficients, make it difficult to interpret the individual impact of features, and sometimes degrade model performance. Removing one of the highly correlated features simplifies the model and improves its stability.
Lasso (L1 Regularization)

Method: Used LassoCV from sklearn.linear_model, which performs cross-validated Lasso regression to select an optimal regularization parameter. Features with non-zero coefficients after Lasso fitting were selected.
Rationale: Lasso not only helps prevent overfitting by penalizing large coefficients but also performs automatic feature selection. It forces the coefficients of less important features to become exactly zero, effectively removing them from the model. This is particularly useful for identifying the most influential features in a linear context.
Random Forest Feature Importance

Method: Used RandomForestRegressor and extracted feature_importances_. Features were then ranked based on these importance scores.
Rationale: Random Forest is a tree-based ensemble method that can capture non-linear relationships and interactions between features. Its feature_importances_ attribute provides a robust measure of how much each feature contributes to the model's predictive power (e.g., by reducing impurity or error). It's a non-parametric method that is less sensitive to multicollinearity compared to linear models for feature importance ranking.
Overall Approach: By combining these methods, we aimed for a robust feature selection strategy that:

Addresses multicollinearity to improve model stability and interpretability.
Leverages a linear model (Lasso) for identifying strong linear predictors.
Utilizes a non-linear model (Random Forest) to capture features that might be important due to complex interactions.

##### Do you think the dataset is imbalanced? Explain Why.

Since this is a regression dataset (predicting Close price), the concept of imbalance is different from classification. In regression, imbalance is observed when the target variable is skewed, meaning most values fall within a narrow range while a few extreme values are rare.

#### 2. Cross- Validation & Hyperparameter Tuning

````python
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# =====================================
# Step 1: Define the base model
# =====================================
model3 = XGBRegressor(
    objective='reg:squarederror',
    random_state=42
)

# =====================================
# Step 2: Define hyperparameter grid
# =====================================
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0]
}

# =====================================
# Step 3: Initialize GridSearchCV
# =====================================
grid_search_xgb = GridSearchCV(
    estimator=model3,
    param_grid=param_grid,
    scoring='r2',
    cv=5,
    n_jobs=-1,
    verbose=1
)

# =====================================
# Step 4: Fit the model
# =====================================

logging.info("Model training started.")

try:
    grid_search_xgb.fit(X_train, y_train)
    print("Model trained successfully.")
except Exception as e:
    logging.info("Model training completed successfully.")
# =====================================
# Step 5: Get best model
# =====================================
best_xgb = grid_search_xgb.best_estimator_
print("Best Hyperparameters:", grid_search_xgb.best_params_)

# =====================================
# Step 6: Predict on test data
# =====================================
y_pred_xgb_tuned = best_xgb.predict(X_test)

# =====================================
# Step 7: Evaluate the model
# =====================================
r2_xgb_tuned = r2_score(y_test, y_pred_xgb_tuned)
rmse_xgb_tuned = np.sqrt(mean_squared_error(y_test, y_pred_xgb_tuned))

print("\nXGBoost Regression Results (After Tuning):")
print(f"R² Score: {r2_xgb_tuned:.4f}")
print(f"RMSE: {rmse_xgb_tuned:.4f}")
````

##### Which hyperparameter optimization technique have you used and why?

For Model-3 (XGBoost), I used GridSearchCV as the hyperparameter optimization technique.

GridSearchCV was chosen because it performs an exhaustive search over all specified hyperparameter combinations and evaluates each combination using cross-validation (in this case, 5-fold cross-validation). This ensures that the model is not just optimized for one split of the data, but performs consistently across multiple subsets, improving generalization.

The main reasons for selecting GridSearchCV are:

It systematically evaluates all possible parameter combinations.

It reduces the risk of overfitting by using cross-validation.

It guarantees finding the best parameters within the defined search space.

It is reliable and easy to interpret, making it suitable for academic and structured project settings.

Although techniques like RandomizedSearchCV or Bayesian Optimization can be more computationally efficient for very large search spaces, GridSearchCV was appropriate here because the parameter grid was moderate in size and allowed precise tuning of key XGBoost parameters such as n_estimators, max_depth, and learning_rate.

In summary, GridSearchCV was used because it provides a structured, thorough, and reliable method for selecting optimal hyperparameters while ensuring strong model generalization performance.
